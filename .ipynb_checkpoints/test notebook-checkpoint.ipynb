```python
import torch
import pandas
import PIL

from PIL import Image
```

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
model.conf = 0.25 # NMS confidence threshold
model.max_det = 100 # maximum number of detections per image
```

```python
im = 'test_images/test.png'
results = model(im)
```

```python
print(results)
```
